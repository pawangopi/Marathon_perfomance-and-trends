```python
import pandas as pd

# Loading the dataset
file_path = file_path = 'C:\\Users\\pawan\\Desktop\\Runners project\\world_marathon_majors.csv'
marathondata = pd.read_csv(file_path, encoding='ISO-8859-1')

marathondata.head()

```

```python
# Checking for null values in the dataset
null_values = marathondata.isnull().sum()
null_values[null_values > 0]

# Remove rows with null values
marathon_data = marathondata.dropna()

marathon_data
```

```python
import matplotlib.pyplot as plt

# Converting the 'time' column to timedelta format using .loc[] to avoid SettingWithCopyWarning
marathon_data.loc[:, 'time_converted'] = pd.to_timedelta(marathon_data['time'].str[-8:])

# Extract the marathon names
marathons = marathon_data['marathon'].unique()

plt.figure(figsize=(12, 8))

for marathon in marathons:
    marathon_subset = marathon_data[marathon_data['marathon'] == marathon]
    plt.plot(marathon_subset['year'], marathon_subset['time_converted'].dt.total_seconds() / 60, label=marathon)

plt.title('Winning Times Over the Years for Different Marathons')
plt.xlabel('Year')
plt.ylabel('Winning Time (Minutes)')
plt.legend()
plt.grid(True)
plt.show()

'''
The chart shows how the winning times of major marathons, like the Boston, Berlin, and NYC marathons, have changed over the years. 
In the early years, especially before the 1980s, the times varied quite a bit, with some races taking much longer to complete than others.
This could be because of less consistent training methods, different race conditions, or fewer professional runners. As time went on, especially
starting in the 1980s, the winning times started to improve and become more consistent across all the major marathons. This indicates that runners 
were getting faster overall, likely due to better training techniques, improved sports science, and more competitive fields.
In recent decades, the winning times for these marathons have become very similar, showing that the level of competition is high across the board. 
All the major marathons are now attracting top-tier athletes, and the races are being run at similar speeds. While there are still slight differences 
in how long it takes to win each marathon, these differences are much smaller than they used to be. This suggests that the sport of marathon running 
has become more standardized and competitive globally, with elite runners consistently pushing the limits of human performance.
'''
```

```python
# Calculate the number of marathon wins by country
country_wins = marathon_data['country'].value_counts()

plt.figure(figsize=(15, 8))
ax = country_wins.plot(kind='bar', color='skyblue')

#  labels on top of the bars
for index, value in enumerate(country_wins):
    ax.text(index, value + 2, str(value), ha='center', fontsize=10)

plt.title('Total Number of Marathon Wins by Country')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.grid(axis='y')
plt.show()

'''# Get the top 10 countries with the most wins
top_countries = country_wins.head(10)
bx = top_countries.plot(kind='bar', color='skyblue')
for index, value in enumerate(top_countries):
    bx.text(index, value + 2, str(value), ha='center', fontsize=10)
# Plot the total number of wins by country using a bar chart
plt.title('Top 10 Countries by Number of Marathon Wins')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()'''
```

```python

# Aggregate the number of wins by year and country
wins_by_year_country = marathon_data.groupby(['year', 'country']).size().unstack(fill_value=0)

# Calculate the cumulative sum of wins over time for each country
cumulative_wins = wins_by_year_country.cumsum()

# Plot the stacked area chart to show country dominance over time
plt.figure(figsize=(15, 8))
cumulative_wins.plot.area(ax=plt.gca(), cmap="tab20", alpha=0.8)
plt.title('Cumulative Marathon Wins by Country Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Number of Wins')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Aggregate the number of wins by year and gender
wins_by_year_gender = marathon_data.groupby(['year', 'gender']).size().unstack(fill_value=0)

# Calculate the cumulative sum of wins over time for each gender
cumulative_wins_gender = wins_by_year_gender.cumsum()

# Plot the cumulative wins over time for each gender
plt.figure(figsize=(12, 6))
cumulative_wins_gender.plot(ax=plt.gca())
plt.title('Cumulative Marathon Wins by Gender Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Number of Wins')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Aggregate the average winning times by year, marathon, and gender
average_times_gender = marathon_data.groupby(['year', 'marathon', 'gender'])['time_converted'].mean().unstack()

# Plot the trends for each marathon per gender
marathons = marathon_data['marathon'].unique()

for marathon in marathons:
    plt.figure(figsize=(12, 6))
    marathon_subset = average_times_gender.xs(marathon, level='marathon')
    
    # Create a new DataFrame to hold the converted values
    marathon_times_in_minutes = marathon_subset.apply(lambda col: col.dt.total_seconds() / 60)
    
    marathon_times_in_minutes.plot(ax=plt.gca())
    plt.title(f'Winning Times Over the Years by Gender: {marathon}')
    plt.xlabel('Year')
    plt.ylabel('Average Winning Time (Minutes)')
    plt.legend(title='Gender')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings

# Filter the data for a specific marathon (e.g., Boston Marathon)
marathon_data_boston = marathon_data[marathon_data['marathon'] == 'Boston']

# Convert 'time' to timedelta
marathon_data_boston.loc[:, 'time_converted'] = pd.to_timedelta(marathon_data_boston['time'].str[-8:])

# Convert the time to minutes for easier analysis
marathon_data_boston['winning_time_minutes'] = marathon_data_boston['time_converted'].dt.total_seconds() / 60

# Set the year as the index (necessary for time series analysis)
marathon_data_boston.set_index('year', inplace=True)

# Visualize the time series data
plt.figure(figsize=(10, 6))
plt.plot(marathon_data_boston.index, marathon_data_boston['winning_time_minutes'], label='Winning Time (Minutes)')
plt.title('Boston Marathon Winning Times Over the Years')
plt.xlabel('Year')
plt.ylabel('Winning Time (Minutes)')
plt.grid(True)
plt.show()

# Split the data into training and testing sets
train_data = marathon_data_boston['winning_time_minutes'][:int(0.8 * len(marathon_data_boston))]
test_data = marathon_data_boston['winning_time_minutes'][int(0.8 * len(marathon_data_boston)):]

# Fit an ARIMA model
model = ARIMA(train_data, order=(5, 1, 0))  # The order (p,d,q) can be tuned for better results
model_fit = model.fit()

# Generate predictions and confidence intervals using get_forecast
forecast_obj = model_fit.get_forecast(steps=len(test_data))
forecast = forecast_obj.predicted_mean
conf_int = forecast_obj.conf_int(alpha=0.05)  # 95% confidence interval

# Plot the actual vs forecasted values with confidence intervals
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, train_data, label='Training Data')
plt.plot(test_data.index, test_data, label='Actual Test Data')
plt.plot(test_data.index, forecast, label='Forecasted Data', color='red')
plt.fill_between(test_data.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='red', alpha=0.3, label='Confidence Interval')
plt.title('Boston Marathon Winning Times Forecast')
plt.xlabel('Year')
plt.ylabel('Winning Time (Minutes)')
plt.legend()
plt.grid(True)
plt.show()
```
